```python
df = spark.read.csv("/Volumes/snitch/default/raw/Snitch_Fashion_Sales_Uncleaned.csv", header=True, inferSchema=True)
```

```python
df.printSchema()
df.count()
```

```python
df.display()
```

```python
from pyspark.sql import functions as F
from pyspark.sql.functions import col

df.select([
    F.sum(F.col(c).isNull().cast("int")).alias(c)
    for c in df.columns
]).display()
```

```python
df.groupBy("Order_ID").count().orderBy("count", ascending=False).display()
```

```python
df.filter(col("Units_Sold") < 0).display()
```

```python
df.select("Profit").distinct().display()
```
